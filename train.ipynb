```python
import os
import yaml
```

```python
def create_dataset_yaml(dataset_path, train_dir, val_dir, class_names, output_file):
    data = {
        'path': dataset_path,
        'train': train_dir,
        'val': val_dir,
        'nc': len(class_names),
        'names': class_names
    }
    
    with open(output_file, 'w') as file:
        yaml.dump(data, file, default_flow_style=False)
    print(f"Created {output_file}")
```

```python
dataset_path = r'/kaggle/input/glasses-segm/data'
train_dir = r'images/train'
val_dir = r'images/val'
class_names = ['glasses', 'lens'] 
output_file = os.path.join('/kaggle/working/', 'dataset.yaml')

create_dataset_yaml(dataset_path, train_dir, val_dir, class_names, output_file)
```

```python
import cv2
import numpy as np
```

```python
input_dir = '/kaggle/input/glasses-segm/data/images/test'
output_dir = '/kaggle/working/test_res'

# Ensure the output directory exists

os.makedirs(output_dir, exist_ok=True)

# Iterate over all files in the input directory
for image_file in os.listdir(input_dir):
    # Construct the full path to the image file
    image_path = os.path.join(input_dir, image_file)

    # Ensure it's an image file (you might want to add more image extensions)
    if image_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        # Load the image
        image = cv2.imread(image_path)

        # Run prediction
        results = model.predict(image)

        # Extract masks
        masks = results[0].masks  # List of masks for detected objects
        
        masks = results[0].masks.data.cpu().numpy()  # Get mask data as a NumPy array
        classes = results[0].boxes.cls.cpu().numpy()  # Get class IDs

        mask_glasses = np.zeros((image.shape[0], image.shape[1]), dtype=np.uint8)
        mask_lenses = np.zeros((image.shape[0], image.shape[1]), dtype=np.uint8)
        
        for mask, cls in zip(masks, classes):
            mask_uint8 = (mask * 255).astype(np.uint8)
            if cls == 0:  # Class ID for glasses
                mask_glasses = np.maximum(mask_glasses, mask_uint8)
            elif cls == 1:  # Class ID for lenses
                mask_lenses = np.maximum(mask_lenses, mask_uint8)
        
        result_mask = mask_glasses - mask_lenses
        
        mask_output_path = os.path.join(output_dir, f"{image_file}_mask.png")
        img_output_path = os.path.join(output_dir, image_file)
    
        cv2.imwrite(mask_output_path, result_mask)
        cv2.imwrite(img_output_path, image)
```
